```python
def load_and_split_datetimes(base_path, node_name):
    # Create the filepath
    filepath = os.path.join(base_path, 'DayAheadPrice', node_name, 'Concatted', 'ConcattedPrices.tsv')
    output_dir = os.path.join(base_path, 'DayAheadPrice', node_name, 'Reformatted')
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, 'ReformattedPrices.tsv')
    
    # Load the TSV file into a DataFrame
    df = pd.read_csv(filepath, sep='\t')
    
    # Check if the MTU column exists
    if 'MTU (CET/CEST)' not in df.columns:
        raise ValueError("Expected column 'MTU (CET/CEST)' not found in the file.")
    
    # Split the 'MTU (CET/CEST)' column into two new columns
    df[['start_datetime', 'end_datetime']] = df['MTU (CET/CEST)'].str.split(' - ', expand=True)
    
    # Optionally convert the datetime strings to datetime objects
    df['start_datetime'] = pd.to_datetime(df['start_datetime'], dayfirst=True)
    df['end_datetime'] = pd.to_datetime(df['end_datetime'], dayfirst=True)
    
    # Extract hour of the day from start_datetime
    df['HourOfDay'] = df['start_datetime'].dt.hour
    
    # Save the reformatted data back to a TSV file
    df.to_csv(output_path, sep='\t', index=False)
    print(f"Data Formatted and written to {output_path}")
    
def concat_price_data(base_path, node_name):
    # Construct the target folder path
    target_folder = os.path.join(base_path, 'DayAheadPrice', node_name, 'Raw')
    
    # List all files in the target folder
    files = [os.path.join(target_folder, f) for f in os.listdir(target_folder) if f.endswith('.csv')]
    
    # Check if there are files to process
    if not files:
        raise ValueError("No CSV files found in the directory.")
    
    # Read the first file to get the header
    initial_df = pd.read_csv(files[0])
    header = initial_df.columns.tolist()
    
    # List to hold all dataframes
    dfs = [initial_df]
    
    # Read each file and verify the header
    for file in files[1:]:
        df = pd.read_csv(file)
        if list(df.columns) != header:
            raise ValueError("CSV file headers do not match.")
        dfs.append(df)
    
    # Concatenate all dataframes into one
    final_df = pd.concat(dfs, ignore_index=True)
    
    # Path for the concatenated output
    output_dir = os.path.join(base_path, 'DayAheadPrice', node_name,'Concatted')
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, 'ConcattedPrices.tsv')
    
    # Write the dataframe to a TSV file
    final_df.to_csv(output_path, sep='\t', index=False)
    print(f"Data concatenated and written to {output_path}")


def convert_to_hourly_array(base_path, node_name):
    # Create the filepath for input and output
    input_path = os.path.join(base_path, 'DayAheadPrice', node_name, 'Reformatted', 'ReformattedPrices.tsv')
    output_dir = os.path.join(base_path, 'DayAheadPrice', node_name, 'Array')
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, 'Array_Price_Regional_Day-Ahead.tsv')
    
    # Load the TSV file into a DataFrame
    df = pd.read_csv(input_path, sep='\t')
    
    # Convert start_datetime from string to datetime if not already
    if df['start_datetime'].dtype == 'object':
        df['start_datetime'] = pd.to_datetime(df['start_datetime'], dayfirst=True,format='ISO8601')
    
    # Pivot the DataFrame to get dates as rows and hours as columns
    df['date'] = df['start_datetime'].dt.date  # Extract the date from start_datetime
    df_pivot = df.pivot_table(index='date', columns='HourOfDay', values='Day-ahead Price [EUR/MWh]', aggfunc='mean')
    
    # Reset index to turn the date index back to a column
    df_pivot.reset_index(inplace=True)
    
    # Save the transformed DataFrame to a TSV file
    df_pivot.to_csv(output_path, sep='\t', index=False)
    print(f"Hourly price array saved to {output_path}")

def add_averages_and_rolling(base_path, node_name):
    # Define paths for input and output
    input_path = os.path.join(base_path, 'DayAheadPrice', node_name, 'Array', 'Array_Price_Regional_Day-Ahead.tsv')
    output_path = os.path.join(base_path, 'DayAheadPrice', node_name, 'Array', 'ArrayStats_Price_Regional_Day-Ahead.tsv')
    
    # Load the DataFrame
    df = pd.read_csv(input_path, sep='\t')
    
    # Calculate the daily average price across all hour columns
    hour_columns = [col for col in df.columns if col.startswith('0') or col.isdigit()]  # Adapt if column naming differs
    df['Daily_Average_Price'] = df[hour_columns].mean(axis=1)
    
    # Calculate the seven-day rolling average of the daily average price
    df['Seven_Day_Rolling_Avg'] = df['Daily_Average_Price'].rolling(window=7, min_periods=1, center=True).mean()
    
    # Save the updated DataFrame
    df.to_csv(output_path, sep='\t', index=False)
    print(f"Updated data saved to {output_path}")

def concat_system_gen_files(base_path, node_name):
    # Define the directory paths
    raw_dir = os.path.join(base_path, 'SystemGeneration', node_name, 'Raw')
    concatted_dir = os.path.join(base_path, 'SystemGeneration', node_name, 'Concatted')
    
    # Create the 'Concatted' directory if it doesn't exist
    os.makedirs(concatted_dir, exist_ok=True)
    
    # Get a list of all CSV files in the 'Raw' directory
    csv_files = [f for f in os.listdir(raw_dir) if f.endswith('.csv')]
    
    # List to store all dataframes
    all_dfs = []
    common_header = None
    
    # Process each CSV file
    for file in csv_files:
        file_path = os.path.join(raw_dir, file)
        df = pd.read_csv(file_path)
        
        # Check if the header matches
        header = tuple(df.columns)
        if common_header is None:
            common_header = header
        elif common_header != header:
            print(f"Header mismatch in file: {file}")
            continue
        
        # Append dataframe to the list
        all_dfs.append(df)
    
    # Concatenate all dataframes
    if all_dfs:
        concatenated_df = pd.concat(all_dfs, ignore_index=True)
        
        # Save the concatenated dataframe as a TSV file
        output_filename = "concatted_system_generation.tsv"
        output_path = os.path.join(concatted_dir, output_filename)
        
        concatenated_df.to_csv(output_path, index=False, sep='\t')
        
        print(f"Concatenation complete. File saved as {output_path}")
    else:
        print("No matching headers found or no files to concatenate.")

def load_and_split_datetimes_SysGen(base_path, node_name):
    # Create the filepath
    filepath = os.path.join(base_path, 'SystemGeneration', node_name, 'Concatted', 'Concatted_System_Generation.tsv')
    output_dir = os.path.join(base_path, 'SystemGeneration', node_name, 'Reformatted')
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, 'ReformattedGeneration.tsv')
    
    # Load the TSV file into a DataFrame
    df = pd.read_csv(filepath, sep='\t')
    
    # Check if the MTU column exists
    if 'MTU (CET/CEST)' not in df.columns:
        raise ValueError("Expected column 'MTU (CET/CEST)' not found in the file.")
    
    # Split the 'MTU (CET/CEST)' column into two new columns
    df[['start_datetime', 'end_datetime']] = df['MTU (CET/CEST)'].str.split(' - ', expand=True)
    
    # Optionally convert the datetime strings to datetime objects
    df['start_datetime'] = pd.to_datetime(df['start_datetime'], dayfirst=True)
    df['end_datetime'] = pd.to_datetime(df['end_datetime'], dayfirst=True)
    
    # Extract hour of the day from start_datetime
    df['HourOfDay'] = df['start_datetime'].dt.hour
    
    # Save the reformatted data back to a TSV file
    df.to_csv(output_path, sep='\t', index=False)
    print(f"Data formatted and written to {output_path}")
```

```python
def convert_sysgen_to_hourly_array(base_path, node_name):
    # Create the filepath for input and output
    input_path = os.path.join(base_path, 'SystemGeneration', node_name, 'Reformatted', 'ReformattedGeneration.tsv')
    output_dir = os.path.join(base_path, 'SystemGeneration', node_name, 'Array')
    os.makedirs(output_dir, exist_ok=True)
    
    # Load the TSV file into a DataFrame
    df = pd.read_csv(input_path, sep='\t')
    
    # Convert start_datetime from string to datetime if not already
    if df['start_datetime'].dtype == 'object':
        df['start_datetime'] = pd.to_datetime(df['start_datetime'], dayfirst=True,format='ISO8601')
    
    # Extract the date from start_datetime
    df['date'] = df['start_datetime'].dt.date
    
    # Define the types and categories
    types = ['Solar', 'Wind Onshore', 'Wind Offshore']
    categories = ['Day Ahead', 'Intraday ', 'Current ']
    
    # Iterate over types and categories to create separate TSV files
    for t in types:
        for c in categories:
            # Create the column name to pivot
            column_name = f'Generation - {t}  [MW] {c}/ BZN|{node_name}'
            
            if column_name in df.columns:
                # Pivot the DataFrame to get dates as rows and hours as columns
                df_pivot = df.pivot_table(index='date', columns='HourOfDay', values=column_name, aggfunc='mean')
                
                # Reset index to turn the date index back to a column
                df_pivot.reset_index(inplace=True)
                
                # Create the output filename
                output_filename = f'Array_Generation_{t.replace(" ", "-")}_{c.replace(" ", "-")}.tsv'
                output_path = os.path.join(output_dir, output_filename)
                
                # Save the transformed DataFrame to a TSV file
                df_pivot.to_csv(output_path, sep='\t', index=False)
                print(f"Hourly array saved to {output_path}")

def array2vector(base_path, node_name):
    # Create the filepath for input and output
    input_path = os.path.join(base_path, 'DayAheadPrice', node_name, 'Array', 'Array_Price_Regional_Day-Ahead.tsv')
    output_dir = os.path.join(base_path, 'DayAheadPrice', node_name, 'Vector')
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, 'Vector_Price_Regional_Day-Ahead.tsv')
    
    # Load the TSV file into a DataFrame
    df = pd.read_csv(input_path, sep='\t')
    
    # Assuming the first column is Date and the rest are values, convert Date column to datetime
    df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0])
    
    # Set the Date column as the index
    df.set_index(df.columns[0], inplace=True)
    
    # Stack the DataFrame to convert it from an array to a vector
    vector_df = df.stack()
    
    # Reset the index to turn the multi-index into columns
    vector_df = vector_df.reset_index()
    
    # Rename the columns appropriately
    vector_df.columns = ['Date', 'Hour', 'Price']
    
    # Save the resulting DataFrame to a TSV file
    vector_df.to_csv(output_path, sep='\t', index=False)
    
    return vector_df
    
```

```python
def fill_gas_prices(file_path, out_path, file_name, start_date, end_date):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(f'{file_path}{file_name}')
    
    # Convert the first column to datetime and set it as the index
    df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0])
    df.set_index(df.columns[0], inplace=True)
    
    # Keep only the first column
    df = df.iloc[:, 0:1]

    # Create a date range from start_date to end_date
    date_range = pd.date_range(start=start_date, end=end_date)
    
    # Reindex the DataFrame to include all dates in the range
    df = df.reindex(date_range)
    
    # Fill missing values using forward fill method
    df.fillna(method='ffill', inplace=True)
    
    # If there are still NaN values at the beginning, use backward fill
    df.fillna(method='bfill', inplace=True)

    os.makedirs(out_path,exist_ok=True)
    
    df.to_csv(f'{out_path}{file_name[:-4]}.tsv', sep= '\t')
              
    return df

def gas_2_power(file_path, out_path, file_name, efficiency):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(f'{file_path}{file_name}',sep='\t')
    
    # Convert the first column to datetime and set it as the index
    df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0])
    df.set_index(df.columns[0], inplace=True)
    
    # Keep only the first column
    df = df.iloc[:, 0:1]
    
    # Divide each value by pp efficiency factor
    df = df / efficiency

    # Ensure the output directory exists
    os.makedirs(out_path, exist_ok=True)
    
    # Save the DataFrame to a CSV file
    df.to_csv(f'{out_path}{file_name}', sep='\t')
              
    return df
```

```python
def load_day_ahead_dataframes(folder_path, node_name):
    # Construct the path to the 'Array' directory inside the node_name directory
    array_path = os.path.join(folder_path, node_name, "Array")
    print(f'loading files in {array_path}')

    names = None
    tmp_names = []
    # List to store the dataframes
    dataframes = []
    
    # Check if the directory exists
    if not os.path.isdir(array_path):
        print(f"Directory does not exist: {array_path}")
        return dataframes
    
    # Loop through each file in the 'Array' directory
    for filename in os.listdir(array_path):
        # Split the filename to check the conditions
        parts = filename.split('_')
        
        
        # Check if the file matches the required pattern
        if len(parts) == 4 and parts[0] == "Array" and parts[-1].startswith("Day-Ahead"):
            # Construct full file path
            file_path = os.path.join(array_path, filename)
            
            # Load the file into a DataFrame and set the first column as a DateTimeIndex
            df = pd.read_csv(file_path, sep='\t', index_col=0, parse_dates=True)
            dataframes.append(df)
            tmp_names.append(parts[2])

    if tmp_names:
        names = tmp_names
    
    return dataframes, names


def compare_dataframes(df_multi, df_single):
    # Ensure both DataFrames have the same index
    if not df_multi.index.equals(df_single.index):
        raise ValueError("Both DataFrames must have the same datetime index.")
    
    # Perform the comparison
    comparison_df = (df_multi.gt(df_single.iloc[:, 0], axis=0)).astype(int)
    return comparison_df

def compare_and_multiply_dataframes(df_multi, df_single):
    # Ensure both DataFrames have the same index
    if not df_multi.index.equals(df_single.index):
        raise ValueError("Both DataFrames must have the same datetime index.")
    
    # Perform the comparison
    comparison_df = (df_multi.gt(df_single.iloc[:, 0], axis=0)).astype(int)
    
    # Multiply each column in the comparison_df by the corresponding value in df_single
    multiplied_df = comparison_df.mul(df_single.iloc[:, 0], axis=0)
    
    return multiplied_df
```

```python
def calculate_seasonal_averages(df):
    # Define summer and winter months
    summer_months = (6, 7, 8)  # June, July, August
    winter_months = (12, 1, 2)  # December, January, February
    
    # Filter data by season
    summer_data = df[df.index.month.isin(summer_months)]
    winter_data = df[df.index.month.isin(winter_months)]
    
    # Calculate averages for each season
    summer_averages = summer_data.mean()
    winter_averages = winter_data.mean()
    
    return summer_averages, winter_averages

def calculate_seasonal_averages2(df):
    # Define months for each season
    seasons = {
        'Spring': (3, 4, 5),  # March, April, May
        'Summer': (6, 7, 8),  # June, July, August
        'Autumn': (9, 10, 11),  # September, October, November
        'Winter': (12, 1, 2)  # December, January, February
    }
    
    # Dictionary to store averages for each season
    seasonal_averages = {}
    for season, months in seasons.items():
        # Filter data by season
        season_data = df[df.index.month.isin(months)]
        # Calculate averages for the season
        seasonal_averages[season] = season_data.mean()
    
    return seasonal_averages


def plot_column_averages(dataframes,title='title',y_axis_label = '€/MWh', x_axis_label = 'Hour_of_Day'):
    # Check if the list is empty
    if not dataframes:
        print("No dataframes provided.")
        return
    
    # Prepare a figure for plotting
    plt.figure(figsize=(10, 6))
    
    # Iterate through the list of DataFrames
    for index, df in enumerate(dataframes):
        # Calculate the mean of each column
        averages = df.mean()
        
        # Plotting the average values
        plt.plot(averages, label=f'DF {index+1}', marker='o')
    
    # Adding labels and title to the plot
    if not title:
        plt.title('Average of Each Column per DataFrame')
    else:
        plt.title(title)
        
    plt.xlabel(x_axis_label)
    plt.ylabel(y_axis_label)
    plt.xticks(rotation=45)  # Rotate column labels for better readability
    plt.legend()  # Add a legend to differentiate the dataframes
    
    # Display the plot
    plt.tight_layout()
    plt.show()

def plot_column_yearly_averages(dataframes, title='title',y_axis_label = '€/MWh',x_axis_label = 'Hour_of_Day'):
    # Check if the list is empty
    if not dataframes:
        print("No dataframes provided.")
        return
    
    # Prepare a figure for plotting
    plt.figure(figsize=(10, 6))
    
    # Iterate through the list of DataFrames
    for index, df in enumerate(dataframes):
        # Check if the DataFrame's index is datetime to allow resampling
        if pd.api.types.is_datetime64_any_dtype(df.index):
            # Resample the DataFrame yearly and calculate the mean for each year
            df_resampled = df.resample('YE').mean()
            # Plot the average values for each year
            for year in df_resampled.index.year:
                yearly_data = df_resampled.loc[df_resampled.index.year == year]
                plt.plot(yearly_data.columns, yearly_data.iloc[0], 
                         label=f'DF {index+1} - {year}', marker='o')
        else:
            print("Index must be datetime to resample by year.")
            return
    
    # Adding labels and title to the plot
    if not title:
        plt.title('Yearly Average of Each Column per DataFrame')
    else:
        plt.title(title)
        
    plt.xlabel(x_axis_label)
    plt.ylabel(y_axis_label)
    plt.xticks(rotation=45)  # Rotate column labels for better readability
    plt.legend(title='Yearly Data', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside the plot
    
    # Display the plot
    plt.tight_layout()
    plt.show()


def plot_averages_and_yearly_averages(dataframes, title='title', y_axis_label='€/MWh', x_axis_label='Hour_of_Day'):
    if not dataframes:
        print("No dataframes provided.")
        return
    
    # Setup the figure and subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))
    
    # First subplot for column averages
    for index, df in enumerate(dataframes):
        averages = df.mean()
        ax1.plot(averages, label=f'DF {index+1}', marker='o')
        
    ax1.set_title(f"{title} - Averages")
    ax1.set_xlabel(x_axis_label)
    ax1.set_ylabel(y_axis_label)
    ax1.tick_params(axis='x', rotation=45)
    ax1.legend()

    # Second subplot for yearly averages
    for index, df in enumerate(dataframes):
        if pd.api.types.is_datetime64_any_dtype(df.index):
            df_resampled = df.resample('YE').mean()
            for year in df_resampled.index.year:
                yearly_data = df_resampled.loc[df_resampled.index.year == year]
                ax2.plot(yearly_data.columns, yearly_data.iloc[0], label=f'DF {index+1} - {year}', marker='o')
        else:
            print("Index must be datetime to resample by year.")
            return

    ax2.set_title(f"{title} - Yearly Averages")
    ax2.set_xlabel(x_axis_label)
    ax2.set_ylabel(y_axis_label)
    ax2.tick_params(axis='x', rotation=45)
    ax2.legend(title='Yearly Data', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Adjust layout and display
    plt.tight_layout()
    plt.show()

def plot_averages_and_yearly_averages(dataframes, title='title', y_axis_label='€/MWh', x_axis_label='Hour_of_Day', series_labels=None):
    if not dataframes:
        print("No dataframes provided.")
        return

    if series_labels is not None and len(series_labels) != len(dataframes):
        raise ValueError("Length of series_labels must match the number of dataframes.")

    # Setup the figure and subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))
    
    # First subplot for column averages
    for index, df in enumerate(dataframes):
        averages = df.mean()
        label = series_labels[index] if series_labels else f"DF {index+1}"
        ax1.plot(averages, label=label, marker='o')
        
    ax1.set_title(f"{title} - Averages")
    ax1.set_xlabel(x_axis_label)
    ax1.set_ylabel(y_axis_label)
    ax1.tick_params(axis='x', rotation=45)
    ax1.legend()

    # Second subplot for yearly averages
    for index, df in enumerate(dataframes):
        if pd.api.types.is_datetime64_any_dtype(df.index):
            df_resampled = df.resample('YE').mean()
            for year in df_resampled.index.year:
                yearly_data = df_resampled.loc[df_resampled.index.year == year]
                label = f"{series_labels[index]} - {year}" if series_labels else f"DF {index+1} - {year}"
                ax2.plot(yearly_data.columns, yearly_data.iloc[0], label=label, marker='o')
        else:
            print("Index must be datetime to resample by year.")
            return

    ax2.set_title(f"{title} - Yearly Averages")
    ax2.set_xlabel(x_axis_label)
    ax2.set_ylabel(y_axis_label)
    ax2.tick_params(axis='x', rotation=45)
    ax2.legend(title='Yearly Data', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Adjust layout and display
    plt.tight_layout()
    plt.show()

def plot_averages_and_yearly_averages(dataframes, title='title', y_axis_label='€/MWh', x_axis_label='Hour_of_Day', series_labels=None):
    if not dataframes:
        print("No dataframes provided.")
        return

    if series_labels is not None and len(series_labels) != len(dataframes):
        raise ValueError("Length of series_labels must match the number of dataframes.")

    # Determine the number of rows needed for the subplots based on the number of dataframes
    num_dfs = len(dataframes)
    fig, axs = plt.subplots(num_dfs, 2, figsize=(20, 6 * num_dfs))  # Adjust the figure size dynamically

    # Iterate over each DataFrame and its corresponding subplot row
    for index, df in enumerate(dataframes):
        # Accessing axes for DataFrames with a potential single subplot situation
        if num_dfs == 1:
            ax1, ax2 = axs[0], axs[1]
        else:
            ax1, ax2 = axs[index, 0], axs[index, 1]
        
        # Plot averages in the first column of subplots
        averages = df.mean()
        label = series_labels[index] if series_labels else f"DF {index+1}"
        ax1.plot(averages, label=label, marker='o')
        ax1.set_title(f"{title} - Averages")
        ax1.set_xlabel(x_axis_label)
        ax1.set_ylabel(y_axis_label)
        ax1.tick_params(axis='x', rotation=45)
        ax1.legend()

        # Plot yearly averages in the second column of subplots
        if pd.api.types.is_datetime64_any_dtype(df.index):
            df_resampled = df.resample('YE').mean()
            for year in df_resampled.index.year:
                yearly_data = df_resampled.loc[df_resampled.index.year == year]
                label = f"{series_labels[index]} - {year}" if series_labels else f"DF {index+1} - {year}"
                ax2.plot(yearly_data.columns, yearly_data.iloc[0], label=label, marker='o')
        else:
            print("Index must be datetime to resample by year.")
            return

        ax2.set_title(f"{title} - Yearly Averages")
        ax2.set_xlabel(x_axis_label)
        ax2.set_ylabel(y_axis_label)
        ax2.tick_params(axis='x', rotation=45)
        ax2.legend(title='Yearly Data', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Adjust layout and display
    plt.tight_layout()
    plt.show()

def plot_averages_and_yearly_averages(dataframes, title='title', y_axis_label='€/MWh', x_axis_label='Hour_of_Day', series_labels=None):
    if not dataframes:
        print("No dataframes provided.")
        return

    if series_labels is not None and len(series_labels) != len(dataframes):
        raise ValueError("Length of series_labels must match the number of dataframes.")

    # Determine the number of rows needed for the subplots based on the number of dataframes
    num_dfs = len(dataframes)
    fig, axs = plt.subplots(num_dfs, 3, figsize=(30, 6 * num_dfs))  # Three columns for different plots

    # Iterate over each DataFrame and its corresponding subplot row
    for index, df in enumerate(dataframes):
        if num_dfs == 1:
            ax1, ax2, ax3 = axs[0], axs[1], axs[2]
        else:
            ax1, ax2, ax3 = axs[index, 0], axs[index, 1], axs[index, 2]
        
        # Plot averages in the first column of subplots
        averages = df.mean()
        label = series_labels[index] if series_labels else f"DF {index+1}"
        ax1.plot(averages, label=label, marker='o')
        ax1.set_title(f"{title} - Averages")
        ax1.set_xlabel(x_axis_label)
        ax1.set_ylabel(y_axis_label)
        ax1.tick_params(axis='x', rotation=45)
        ax1.legend()

        # Plot yearly averages in the second column of subplots
        if pd.api.types.is_datetime64_any_dtype(df.index):
            df_resampled = df.resample('YE').mean()
            for year in df_resampled.index.year:
                yearly_data = df_resampled.loc[df_resampled.index.year == year]
                label = f"{series_labels[index]} - {year}" if series_labels else f"DF {index+1} - {year}"
                ax2.plot(yearly_data.columns, yearly_data.iloc[0], label=label, marker='o')
        else:
            print("Index must be datetime to resample by year.")
            return

        ax2.set_title(f"{title} - Yearly Averages")
        ax2.set_xlabel(x_axis_label)
        ax2.set_ylabel(y_axis_label)
        ax2.tick_params(axis='x', rotation=45)
        ax2.legend()

        # Get summer and winter averages from the helper function
        summer_averages, winter_averages = calculate_seasonal_averages(df)
        # Plot seasonal averages in the third column of subplots
        ax3.plot(summer_averages, label=f'{series_labels[index] if series_labels else "DF "+str(index+1)} - Summer', marker='o', color='red')
        ax3.plot(winter_averages, label=f'{series_labels[index] if series_labels else "DF "+str(index+1)} - Winter', marker='o', color='blue')
        ax3.set_title(f"{title} - Seasonal Averages")
        ax3.set_xlabel(x_axis_label)
        ax3.set_ylabel(y_axis_label)
        ax3.tick_params(axis='x', rotation=45)
        ax3.legend()

    # Adjust layout and display
    plt.tight_layout()
    plt.show()

def plot_averages_and_yearly_averages2(dataframes, title='title', y_axis_label='€/MWh', x_axis_label='Hour_of_Day', series_labels=None):
    if not dataframes:
        print("No dataframes provided.")
        return

    if series_labels is not None and len(series_labels) != len(dataframes):
        raise ValueError("Length of series_labels must match the number of dataframes.")

    # Determine the number of rows needed for the subplots based on the number of dataframes
    num_dfs = len(dataframes)
    # Four columns for different plots including a new one for each season
    fig, axs = plt.subplots(num_dfs, 4, figsize=(40, 6 * num_dfs)) 

    # Iterate over each DataFrame and its corresponding subplot row
    for index, df in enumerate(dataframes):
        if num_dfs == 1:
            ax1, ax2, ax3, ax4 = axs[0], axs[1], axs[2], axs[3]
        else:
            ax1, ax2, ax3, ax4 = axs[index, 0], axs[index, 1], axs[index, 2], axs[index, 3]

        # Plot averages in the first column of subplots
        averages = df.mean()
        label = series_labels[index] if series_labels else f"DF {index+1}"
        ax1.plot(averages, label=label, marker='o')
        ax1.set_title(f"{title} - Averages")
        ax1.set_xlabel(x_axis_label)
        ax1.set_ylabel(y_axis_label)
        ax1.tick_params(axis='x', rotation=45)
        ax1.legend()

        # Plot yearly averages in the second column of subplots
        if pd.api.types.is_datetime64_any_dtype(df.index):
            df_resampled = df.resample('YE').mean()
            for year in df_resampled.index.year:
                yearly_data = df_resampled.loc[df_resampled.index.year == year]
                label = f"{series_labels[index]} - {year}" if series_labels else f"DF {index+1} - {year}"
                ax2.plot(yearly_data.columns, yearly_data.iloc[0], label=label, marker='o')
        else:
            print("Index must be datetime to resample by year.")
            return

        ax2.set_title(f"{title} - Yearly Averages")
        ax2.set_xlabel(x_axis_label)
        ax2.set_ylabel(y_axis_label)
        ax2.tick_params(axis='x', rotation=45)
        ax2.legend()

        # Get seasonal averages from the helper function
        seasonal_averages = calculate_seasonal_averages2(df)
        # Plot seasonal averages in the fourth column of subplots
        for season, averages in seasonal_averages.items():
            ax4.plot(averages, label=f'{series_labels[index] if series_labels else "DF "+str(index+1)} - {season}', marker='o')

        ax4.set_title(f"{title} - Seasonal Averages")
        ax4.set_xlabel(x_axis_label)
        ax4.set_ylabel(y_axis_label)
        ax4.tick_params(axis='x', rotation=45)
        ax4.legend()

    # Adjust layout and display
    plt.tight_layout()
    plt.show()

def plot_averages_and_yearly_averages2(dataframes, title='title', y_axis_label='€/MWh', x_axis_label='Hour_of_Day', series_labels=None):
    if not dataframes:
        print("No dataframes provided.")
        return

    if series_labels is not None and len(series_labels) != len(dataframes):
        raise ValueError("Length of series_labels must match the number of dataframes.")

    # Determine the number of rows needed for the subplots based on the number of dataframes
    num_dfs = len(dataframes)
    # Three columns for different plots
    fig, axs = plt.subplots(num_dfs, 3, figsize=(30, 6 * num_dfs)) 

    # Iterate over each DataFrame and its corresponding subplot row
    for index, df in enumerate(dataframes):
        if num_dfs == 1:
            ax1, ax2, ax3 = axs[0], axs[1], axs[2]
        else:
            ax1, ax2, ax3 = axs[index, 0], axs[index, 1], axs[index, 2]

        # Plot averages in the first column of subplots
        averages = df.mean()
        label = series_labels[index] if series_labels else f"DF {index+1}"
        ax1.plot(averages, label=label, marker='o')
        ax1.set_title(f"{title} - Averages")
        ax1.set_xlabel(x_axis_label)
        ax1.set_ylabel(y_axis_label)
        ax1.tick_params(axis='x', rotation=45)
        ax1.legend()

        # Plot yearly averages in the second column of subplots
        if pd.api.types.is_datetime64_any_dtype(df.index):
            df_resampled = df.resample('YE').mean()
            for year in df_resampled.index.year:
                yearly_data = df_resampled.loc[df_resampled.index.year == year]
                label = f"{series_labels[index]} - {year}" if series_labels else f"DF {index+1} - {year}"
                ax2.plot(yearly_data.columns, yearly_data.iloc[0], label=label, marker='o')
        else:
            print("Index must be datetime to resample by year.")
            return

        ax2.set_title(f"{title} - Yearly Averages")
        ax2.set_xlabel(x_axis_label)
        ax2.set_ylabel(y_axis_label)
        ax2.tick_params(axis='x', rotation=45)
        ax2.legend()

        # Get seasonal averages from the helper function
        seasonal_averages = calculate_seasonal_averages2(df)
        # Plot seasonal averages in the third column of subplots
        for season, averages in seasonal_averages.items():
            ax3.plot(averages, label=f'{series_labels[index] if series_labels else "DF "+str(index+1)} - {season}', marker='o')

        ax3.set_title(f"{title} - Seasonal Averages")
        ax3.set_xlabel(x_axis_label)
        ax3.set_ylabel(y_axis_label)
        ax3.tick_params(axis='x', rotation=45)
        ax3.legend()

    # Adjust layout and display
    plt.tight_layout()
    plt.show()

```

```python


def run_linear_regression(dependent_var_list, independent_vars_list, include_constant=True, include_squares=False):
    # Extract the single DataFrame for the dependent variable
    dependent_df = dependent_var_list[0]

    # Flatten the dependent DataFrame if it's not already a series
    if len(dependent_df.columns) > 1:
        dependent_series = dependent_df.stack()
    else:
        dependent_series = dependent_df.iloc[:, 0]

    # Prepare a list for all flattened independent DataFrames
    independent_data = []

    # Flatten each independent variable DataFrame and store it
    for df in independent_vars_list:
        if len(df.columns) > 1:
            flattened_df = df.stack()
            independent_data.append(flattened_df)
        else:
            independent_data.append(df.iloc[:, 0])

    # Create a DataFrame from the list of independent variable series
    independent_df = pd.concat(independent_data, axis=1)

    # Make sure indices are aligned and there are no missing values
    data = pd.concat([dependent_series, independent_df], axis=1).dropna()

    # Separate dependent and independent variables
    Y = data.iloc[:, 0]
    X = data.iloc[:, 1:]

    # Optionally add a constant to the model
    if include_constant:
        X = sm.add_constant(X)

    # Optionally include the squares of independent variables
    if include_squares:
        for col in X.columns[1:]:  # Skip the constant if it exists
            X[f'{col}^2'] = X[col] ** 2

    # Run the regression
    model = sm.OLS(Y, X)
    results = model.fit()

    # Return regression results
    return results

def multiply_coefficients_with_columns(results, independent_vars_list, include_constant=True, include_squares=False):
    coefficients = results.params

    # Prepare a list for all flattened independent DataFrames
    independent_data = []

    # Flatten each independent variable DataFrame and store it
    for df in independent_vars_list:
        df = df.ffill()
        print(len(df))
        if len(df.columns) > 1:
            flattened_df = df.stack()
            independent_data.append(flattened_df)
        else:
            independent_data.append(df.iloc[:, 0])

    # Create a DataFrame from the list of independent variable series
    independent_df = pd.concat(independent_data, axis=1, join='inner')
    
    # Optionally include the squares of independent variables
    if include_squares:
        for col in independent_df.columns[0:]:
            independent_df[f'{col}^2'] = independent_df[col] ** 2
    
    # Multiply each column by its respective coefficient
    multiplied_df = independent_df * coefficients
    # Optionally add a constant to the DataFrame
    
    if include_constant:
        multiplied_df = sm.add_constant(multiplied_df)
    else:
        multiplied_df = multiplied_df.drop(columns='const')

    return multiplied_df
```
